```python
from parcs.cdag.graph_objects import Graph
from parcs.graph_builder.randomizer import ParamRandomizer
from matplotlib import pyplot as plt

rnd = ParamRandomizer(graph_dir='gdf.yml', guideline_dir='guideline.yml')
nodes, edges = rnd.get_graph_params()
graph = Graph(nodes, edges)
data = graph.sample(100)

plt.figure(figsize=(8, 8))
plt.scatter(data.X, data.Y, c=data.Z, s=70)
plt.xlabel('X', fontsize=25)
plt.ylabel('Y', fontsize=25)
plt.savefig('fig3.png')
plt.show()
```
